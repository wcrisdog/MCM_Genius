```python
import plotly
import chart_studio.plotly as py
import pandas as pd
import numpy as np
import cufflinks as cf
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
%matplotlib inline
init_notebook_mode(connected=True)
cf.go_offline()
```

```python
df = pd.DataFrame(np.random.randn(100,4),columns='A B C D'.split())
df.head()
```

```python
df2 = pd.DataFrame({'Category':['A','B','C'],'Values':[32,43,50]})
```

```python
df.iplot()
```

```python
df.iplot(kind='scatter',x='A',y='B',mode='markers',size=10)  #散点图，放进x和y的数据,点图，点大小
```

```python
print(df2)
df2.iplot(kind='bar',x='Category',y='Values')   #根据df2的类型写出柱状图
```

```python
print(df)
df.count().iplot(kind='bar')    #df很乱，用count() or sum()画出柱状图
```

```python
df.iplot(kind='box',x='A',y='B')    #忘记叫什么的，完整信息box图
```

```python
df3 = pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,20,10],'z':[5,4,3,2,1]})
print(df3)
df3.iplot(kind='surface',colorscale='rdylbu') #三维立体图，可以自己设置color pattern，可查
```

```python
df['A'].iplot(kind='hist',bins=50)    #频率直方图,某一行
#也可以把‘A’去掉，看一整个的直方图
```

```python
pd.np = np
df[['A','B']].iplot(kind='spread')
#股价图，但有bug，不知道为什么

# Works now, 应该是cufflinks的代码实现用了老版本的pandas 那时的pandas有pd.np，现在没了就得单独调用np
```

```python
df.iplot(kind='bubble',x='A',y='B',size='C')   #泡泡图，用于GDP等，数据见df
```
