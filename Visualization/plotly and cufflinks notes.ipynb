```python
import plotly
import chart_studio.plotly as py
import pandas as pd
import numpy as np
import cufflinks as cf
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
%matplotlib inline
init_notebook_mode(connected=True)
cf.go_offline()
```

```python
df = pd.DataFrame(np.random.randn(100,4),columns='A B C D'.split())
df.head()
```

```python
df2 = pd.DataFrame({'Category':['A','B','C'],'Values':[32,43,50]})
```

```python
df.iplot()
```

```python
df.iplot(kind='scatter',x='A',y='B',mode='markers',size=10)  #xy,
```

```python
print(df2)
df2.iplot(kind='bar',x='Category',y='Values')   #df2
```

```python
print(df)
df.count().iplot(kind='bar')    #dfcount() or sum()
```

```python
df.iplot(kind='box',x='A',y='B')    #box
```
